# svm实现概率预测：重要接口probability，接口predict_proba以及decision_function

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
# 使用样本不均衡的模型
class_1 = 500 # 类别1有500个样本
class_2 = 50 # 类别2有50个样本
centers = [[0.0,0.0],[2.0,2.0]] # 设定两个类别的中心
clusters_std = [1.5,0.5] # 设定两个类别的方差 通常来说样本量大的类别会更松散
x,y = make_blobs(n_samples=[class_1,class_2],
                centers = centers,
                cluster_std = clusters_std,
                random_state = 0,
                shuffle = False)
```

```python
x
```

```python
y
```

```python
# 画出数据的样子
plt.figure(figsize=(12,8))
plt.scatter(x[:,0],x[:,1],c=y,cmap="rainbow",s=10)
```

```python
from sklearn import svm
# 实例化支持向量机分类器，核函数使用linear
clf_proba = svm.SVC(kernel='linear',C=1.0,probability=True).fit(x,y)
```

```python
clf_proba.predict_proba(x).shape
```

```python
# decision_function 生成的是点到决策边界的距离
clf_proba.decision_function(x).shape
```
