```python
from sklearn.datasets import make_blobs
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
# 使用make_blobs创建一组数据,第一个参数是数据点的数量，第二个参数是中心个数
# sklearn.datasets包中关于make的包是用来制造数据的，他们的返回值类型是元祖
# 返回的元祖中的第一项是数据的特征值，第二个参数是数据的标签值，后面会用这两个特征值来画图
# 用标签值作为图中点的颜色值，便于区分颜色。cluster_std表达的是簇的方差
data = make_blobs(n_samples=50,centers=2,random_state=0,cluster_std=0.6)
```

```python
x,y = data
```

```python
# 画出散点图
plt.figure(figsize=(6,6))
# scatter用于画散点图，前两个参数是两个数组，分别是数据点的x坐标的集合和y坐标的集合
# 这里的颜色使用y的值，可以更好的分类。s是点的尺寸，这里设置为50，cmap表示所采用的点的颜色
plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='rainbow')
# 横纵坐标均显示为空,不显示任何内容
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# 开始画决策边界的过程，这里要理解函数counter的意思
# counter是专门用于绘制等高线的函数
plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='rainbow')
ax = plt.gca() # 获取当前的子图，如果不存在就创建子图 这里因为我们已经有了一个散点图，因此我们的ax就是我们的散点图
```

```python
# 制作网格
# 思路：先获取平面上的最大值和最小值，也就是获取两条坐标轴的范围
# 将两个轴分别分为对等数量的点，然后就获得了相对数量的点

# 1. 获取平面上两条坐标轴的最大值和最小值
xlim = ax.get_xlim()
ylim = ax.get_ylim()
```
